```python
from collections import Counter

nlp = spacy.load("en_core_web_sm")

def extract_entities(text):
    doc = nlp(text)
    return [ent.label_ for ent in doc.ents]

df['entities'] = df['content'].apply(extract_entities)

df['org_mentions'] = df['entities'].apply(lambda ents: ents.count("ORG"))
df['place_mentions'] = df['entities'].apply(lambda ents: ents.count("GPE"))
```

```python
df['date_mentions'] = df['entities'].apply(lambda ents: ents.count("DATE"))
```

```python
df['time_mentions'] = df['entities'].apply(lambda ents: ents.count("TIME"))
df['date_mentions'] = df['entities'].apply(lambda ents: ents.count("DATE"))
df['loc_mentions'] = df['entities'].apply(lambda ents: ents.count("LOC"))
df['money_mentions'] = df['entities'].apply(lambda ents: ents.count("MONEY"))
df['percent_mentions'] = df['entities'].apply(lambda ents: ents.count("PERCENT"))
df['product_mentions'] = df['entities'].apply(lambda ents: ents.count("PRODUCT"))
df['event_mentions'] = df['entities'].apply(lambda ents: ents.count("EVENT"))
df['law_mentions'] = df['entities'].apply(lambda ents: ents.count("LAW"))
df['language_mentions'] = df['entities'].apply(lambda ents: ents.count("LANGUAGE"))
df['art_mentions'] = df['entities'].apply(lambda ents: ents.count("WORK_OF_ART"))
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
from wordcloud import STOPWORDS
custom_stopwords = set(STOPWORDS)
custom_stopwords.update(['rt', 'https', 'co', 'amp','hyperlink','s'])
```

```python
import re

text = ' '.join(df['content'].dropna().astype(str).tolist())

text = text.lower()
text = re.sub(r'[^\w\s]', '', text)
text = re.sub(r'\s+', ' ', text).strip()
```

```python
import unicodedata

# Remove all non-printable characters
def clean_text(txt):
    txt = unicodedata.normalize('NFKD', txt)
    txt = ''.join(c for c in txt if c.isprintable())
    return txt

text = clean_text(text)
```

```python
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    stopwords=custom_stopwords
).generate(text)
```

```python
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("WordCloud of Tweet Content")
plt.show()
```

```python
import numpy as np
from PIL import Image
mask = np.array(Image.open('/content/image-from-rawpixel-id-4092679-png.png'))

# Custom stopwords
custom_stopwords = set(STOPWORDS)
custom_stopwords.update(['rt', 'https', 'co', 'amp','hyperlink','s'])

# Clean text
text = ' '.join(df['content'].dropna().astype(str).tolist()).lower()
text = re.sub(r'[^\w\s]', '', text)
text = re.sub(r'\s+', ' ', text).strip()
text = unicodedata.normalize('NFKD', text)
text = ''.join(c for c in text if c.isprintable())

# Generate wordcloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    stopwords=custom_stopwords,
    mask=mask,
    contour_color='black',
    contour_width=1
).generate(text)

# Display
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("WordCloud with Custom Shape")
plt.show()
```
